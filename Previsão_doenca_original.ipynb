```python
# Importar principais bibliotecas para auxiliar na leitura, manipulação, tratamento e análise dos dados
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
treino = pd.read_csv('/content/Training.csv') # Usa biblioteca pandas para abrir arquivo que contem os dados de treino
test = pd.read_csv('/content/Testing.csv') # Arquivo para testar
#Base dos dados: https://www.kaggle.com/datasets/tejpal123/human-disease-prediction-dataset
```

```python
treino = treino.drop('Unnamed: 133',axis=1) # Eliminar a coluna Unnamed: 133 - coluna sem dados
```

```python
#Realiza separação do dados de treino e teste
X_train, y_train = treino.drop('prognosis', axis = 1), treino['prognosis']
X_test, y_test = test.drop('prognosis', axis = 1), test['prognosis']
```

```python
# Aprendizado de máquina utilizado para realizar predições
log = LogisticRegression()
dtr = DecisionTreeClassifier()
rfc = RandomForestClassifier()
svc = SVC(C=0.9)
mnb = MultinomialNB()
knn = KNeighborsClassifier(n_neighbors=1)
xgb = GradientBoostingClassifier()
rna = MLPClassifier( hidden_layer_sizes=(100, 50, 20),activation='logistic')

models = [log, dtr, rfc, svc, mnb, knn, xgb, rna]
names = ["Regressão Logística", "Árvores de decisão", "Random Forest", "Support vector machine", "Multinomial Naive Bayes", "k-Nearest Neighbors (k-NN)", "Gradient Boosting Machines (GBM)",
         "Redes Neurais Artificiais", ]
```

```python
def training(model, name):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    score = accuracy_score(pred, y_test)
    print(name, ": {}%".format(round(score*100, 2)))
    report = classification_report(pred, y_test)
    print(report)

for i, j in zip(models, names):
    print('*'*40)
    training(i, j)
    print()
```
